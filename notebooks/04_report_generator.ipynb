# 04 — Report Generator

This notebook compiles the findings and charts from the previous notebooks into a single PDF report (`recommendation.pdf`) using the `reportlab` library.

The report is structured for a non-technical audience (e.g. a business manager) — it leads with an executive summary, presents each finding with its supporting chart, and ends with concrete recommendations and expected impact.

**Steps:**
1. Import `reportlab` components
2. Define text styles (title, heading, body)
3. Build the report content (summary, findings, charts)
4. Add recommendations and generate the PDF

```python
from reportlab.lib.pagesizes import A4
from reportlab.lib import colors
from reportlab.lib.units import cm
from reportlab.platypus import (SimpleDocTemplate, Paragraph, Spacer, 
                                 Image, Table, TableStyle, HRFlowable)
from reportlab.lib.styles import getSampleStyleSheet, ParagraphStyle
from reportlab.lib.enums import TA_CENTER, TA_LEFT
import os

print("✅ Libraries imported!")
```

## Step 2 — Define Styles

We define custom styles for title, heading, and body text to keep the PDF visually consistent.  
Separating style definitions from content makes it easy to update the look of the entire report from one place.

```python
# Set up document
doc = SimpleDocTemplate(
    "../report/recommendation.pdf",
    pagesize=A4,
    rightMargin=2*cm,
    leftMargin=2*cm,
    topMargin=2*cm,
    bottomMargin=2*cm
)

# Set up styles
styles = getSampleStyleSheet()

title_style = ParagraphStyle(
    "CustomTitle",
    parent=styles["Title"],
    fontSize=24,
    textColor=colors.HexColor("#2c3e50"),
    spaceAfter=0.5*cm,
    alignment=TA_CENTER
)

heading_style = ParagraphStyle(
    "CustomHeading",
    parent=styles["Heading1"],
    fontSize=14,
    textColor=colors.HexColor("#2980b9"),
    spaceBefore=0.5*cm,
    spaceAfter=0.3*cm
)

body_style = ParagraphStyle(
    "CustomBody",
    parent=styles["Normal"],
    fontSize=11,
    leading=16,
    spaceAfter=0.3*cm
)

bullet_style = ParagraphStyle(
    "CustomBullet",
    parent=styles["Normal"],
    fontSize=11,
    leading=16,
    leftIndent=0.5*cm,
    spaceAfter=0.2*cm
)

print("✅ Styles defined!")
```

## Step 3 — Build Report Content

We assemble the report content as a list of `reportlab` flowable elements (paragraphs, images, spacers).  
Each finding section follows the same structure: a text description of the finding, followed by the chart that supports it.  
This mirrors the "claim → evidence" structure used in professional business reports.

```python
# Container for all report elements
content = []

# ── TITLE ──────────────────────────────────────────
content.append(Paragraph("Superstore Sales Analysis", title_style))
content.append(Paragraph("Business Recommendation Report", title_style))
content.append(HRFlowable(width="100%", thickness=2, 
                           color=colors.HexColor("#2980b9")))
content.append(Spacer(1, 0.5*cm))

# ── EXECUTIVE SUMMARY ──────────────────────────────
content.append(Paragraph("Executive Summary", heading_style))
content.append(Paragraph(
    "This report analyzes Superstore sales data across four regions to identify "
    "profit inefficiencies and provide actionable business recommendations. "
    "The analysis reveals that the Central region is significantly underperforming "
    "due to aggressive discounting practices.", body_style))

content.append(Spacer(1, 0.3*cm))

# ── KEY FINDINGS ───────────────────────────────────
content.append(Paragraph("Key Findings", heading_style))

content.append(Paragraph(
    "Finding 1 — Central has the lowest profit margin", heading_style))
content.append(Paragraph(
    "Central region generates $501K in sales but achieves only 7.92% profit margin "
    "— nearly half of the West region's 14.94%, despite having higher sales than South.",
    body_style))
content.append(Image("../report/chart1_region_margin.png", 
                      width=14*cm, height=8*cm))
content.append(Spacer(1, 0.3*cm))

content.append(Paragraph(
    "Finding 2 — Central over-discounts compared to other regions", heading_style))
content.append(Paragraph(
    "Central's average discount of 24% is more than double West's 11%. "
    "Nearly 30% of Central's orders are discounted above 30% — "
    "a threshold at which orders typically become unprofitable.",
    body_style))
content.append(Image("../report/chart2_region_discount.png", 
                      width=14*cm, height=8*cm))
content.append(Spacer(1, 0.3*cm))

content.append(Paragraph(
    "Finding 3 — 7 sub-categories are losing money in Central", heading_style))
content.append(Paragraph(
    "Furniture is the most problematic category — 3 out of 4 sub-categories "
    "generate negative profit in Central. Furnishings alone loses $3,906 "
    "despite $15,254 in sales.",
    body_style))
content.append(Image("../report/chart3_central_subcategory.png", 
                      width=14*cm, height=8*cm))
content.append(Spacer(1, 0.3*cm))

content.append(Paragraph(
    "Finding 4 — High discounts directly cause negative profit", heading_style))
content.append(Paragraph(
    "Orders with discounts above 30% are almost universally unprofitable "
    "across all categories in Central region.",
    body_style))
content.append(Image("../report/chart4_discount_vs_profit.png", 
                      width=14*cm, height=8*cm))
content.append(Spacer(1, 0.5*cm))

print("✅ Content built!")
```

## Step 4 — Add Recommendations & Generate PDF

The recommendations are presented in a table format with three columns: recommendation number, action description, and expected business impact.  
Using a table makes it easy for the reader to scan and prioritise actions at a glance.

Running this cell outputs `recommendation.pdf` to the `../report/` folder.

```python
# ── RECOMMENDATIONS ────────────────────────────────
content.append(HRFlowable(width="100%", thickness=1,
                           color=colors.HexColor("#2980b9")))
content.append(Paragraph("Business Recommendations", heading_style))

recommendations = [
    ["#", "Recommendation", "Expected Impact"],
    ["1", "Cap discounts at 15% in Central region\nDiscounts above 20% consistently\nproduce negative profit", 
         "Estimated margin improvement\nfrom 7.92% → 12%+"],
    ["2", "Eliminate deep discounts on Furniture\nFurnishings (40% avg discount) and\nTables (26%) are top loss makers",
         "Recover $7,000+ in lost\nprofit from Furniture alone"],
    ["3", "Review Binders pricing strategy\n$56K in sales but losing money\ndue to 51% average discount",
         "Turn $1,044 loss into\npotential profit"],
]

table = Table(recommendations, colWidths=[1.5*cm, 9*cm, 6*cm])
table.setStyle(TableStyle([
    # Header row
    ("BACKGROUND", (0,0), (-1,0), colors.HexColor("#2980b9")),
    ("TEXTCOLOR", (0,0), (-1,0), colors.white),
    ("FONTNAME", (0,0), (-1,0), "Helvetica-Bold"),
    ("FONTSIZE", (0,0), (-1,0), 11),
    ("ALIGN", (0,0), (-1,-1), "LEFT"),
    ("VALIGN", (0,0), (-1,-1), "MIDDLE"),
    ("PADDING", (0,0), (-1,-1), 8),
    # Alternating row colors
    ("BACKGROUND", (0,1), (-1,1), colors.HexColor("#eaf4fb")),
    ("BACKGROUND", (0,2), (-1,2), colors.white),
    ("BACKGROUND", (0,3), (-1,3), colors.HexColor("#eaf4fb")),
    # Grid
    ("GRID", (0,0), (-1,-1), 0.5, colors.HexColor("#bdc3c7")),
    ("FONTSIZE", (0,1), (-1,-1), 10),
]))

content.append(Spacer(1, 0.3*cm))
content.append(table)
content.append(Spacer(1, 0.5*cm))

# ── CONCLUSION ─────────────────────────────────────
content.append(HRFlowable(width="100%", thickness=1,
                           color=colors.HexColor("#2980b9")))
content.append(Paragraph("Conclusion", heading_style))
content.append(Paragraph(
    "The Central region has strong sales volume but is undermined by aggressive "
    "discounting that erodes profitability. By capping discounts at 15% and "
    "eliminating deep discounts on Furniture, the Central region could realistically "
    "improve its profit margin from 7.92% to above 12% — bringing it closer to "
    "the West region's performance benchmark of 14.94%.",
    body_style))

content.append(Spacer(1, 0.5*cm))
content.append(HRFlowable(width="100%", thickness=2,
                           color=colors.HexColor("#2c3e50")))
content.append(Spacer(1, 0.2*cm))
content.append(Paragraph(
    "Analysis conducted using SQL (SQLite3), Python (Pandas, Matplotlib, Seaborn)",
    ParagraphStyle("footer", parent=styles["Normal"], 
                   fontSize=9, textColor=colors.grey,
                   alignment=TA_CENTER)))

# ── GENERATE PDF ───────────────────────────────────
doc.build(content)
print("✅ recommendation.pdf generated successfully!")
```
